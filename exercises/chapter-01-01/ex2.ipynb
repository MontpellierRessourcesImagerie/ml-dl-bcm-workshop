# 2 Clustering


In the exercises in this chapter you will first use a visualization of the k-means
clustering that helps to understand how it works. You will then apply the k-
means clustering as a machine learning method to segment plants based on their
color and than on features of the scratch-assay-images.

## 2.1 k-means clustering

Open the file ``Visualizing K-Means Clustering.html`` from the folder ``01-k-means-visualization``. The applet allows you to see the k-means clustering at work. You can select different strategies for the initial mean-values and different data–sets. You can follow the k-means algorithm step by step. Run the k-means algorithm with different initialization strategies and on different data–sets and answer the following questions.

1. What influence do you think the selection of the initial mean values has on the clustering result ?
1. Does the k-means algorithm always converge, i.e. does it always come to a point where further iterations do not change the result anymore?
1. Is the result of the clustering always the same, no matter what the choice of the initial means is?
1. Does the k-means clustering work as expected on the Smiley Face data–set? What is the reason it does or does not work as expected?



### 2.2 Using K-means clustering for color segmentation

Copy the scripts [k-means_segmentation.py](../../scripts/k-means_segmentation.py) and [normalize_feature_stack.py](../../scripts/normalize_feature_stack.py) into a subfolder of the plugins folder of your FIJI and restart FIJI.

Images of Arabidopsis thaliana are taken in regular intervals. The aim is to measure the speed with which the area of the rosette of leaves augments under different conditions and treatments. You will use k-means clustering to segment the rosettes of the plants. You will create a color-stack using an appropriate color-space (RGB, HSB or Lab) and use the k-means segmentation tool.

1. What is the class of the color values belonging to the plant?
1. What is the class of the color values belonging to the small red reference object in the form of a ball?
1. Create a selection (roi) from the result of the pixel classification and display it on the original input image. Transfer the selection to the original input image (``restore selection`` or use the roi-manager). How good is the selection?
1. Measure the area of the surface of the plant.
1. Check the log–window. The clustering algorithm has written the k mean-vectors into the log. What are their values?
1. Apply the saved k-means classifier on a different image from the plant-series.

## 2.3 Using k-means clustering on other features

We can use the k-means clustering not only with colors but also with other
features.

1) Open an image from scratch-assay-folder. Create two or three feature images from the input image for example by using FeatureJ or other FIJI filters. Use appropriate sigmas when creating the features. Use ``Image>Color>Merge Channels...`` to create a stack of the feature images. Run the k-means clustering on the stack. This time you might want to check the ``normalize`` option. What is the area of the gap?
