## 1.2 Machine Learning

You will now use machine learning to do the segmentation of the gap-area in the scratch assay analysis.

1. Open some of the images from one of the subfolders of ``data_ex01-scratch-assay``. Build a stack of training images from them and make sure that the stack is a time-series of 2D images.
1. Open the stack in labkit.

3. Switch labkit from 3D mode to ``slice by sclice``-mode. Familiarize yourself with the navigation in labkit (panning, zooming, reset view, ...)
1. Add a classifier and select the features and sigmas you want to use.
1. Train the classifier to predict the classes ``gap`` and ``tissue``. Do as many cycles of annotation and running ``Train Classifier`` as necessary.
1. Open the segmentation results in FIJI.
1. Duplicate the image to get rid of the view and select the label of interest, the gap.
1. Measure the area of the gap on one of the result images.
1. Run the labkit classifier in batch mode on the folder containing the scratch assay images.
1. Check the results of the batch processing in FIJI.
