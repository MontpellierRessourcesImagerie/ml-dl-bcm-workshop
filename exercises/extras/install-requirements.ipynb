# Install requirements

## I. General

- A laptop with a decent amount of RAM (16Go bare minimum).
- We strongly recommend everybody to take a **mouse**.
- Make some **free space on your disk**, the images that we will manipulate should be present locally on your disk and some data will be generated over the course of the workshop.
- If you have the possibility to get an **Eduroam** access, activate it before coming. An internet connection is required anyways.
- If you have an **NVidia graphics card**, try to have your drivers working and up-to-date.


## II. Datasets

- You should download the dataset that we are going to use in the different exercises. 
- It is heavy and **there is no guest internet** access.
- You can download the dataset from sDrive: [https://sdrive.cnrs.fr/s/L4Z6mKTcaRBCeZc](https://sdrive.cnrs.fr/s/L4Z6mKTcaRBCeZc).


# III. Python

### 1. MiniConda

- To avoid collisions in Python modules, we will use virtual environments.
- If you are more familiar with a virtual environments manager different of MiniConda, feel free to use it, but the examples in the workshop will be with MiniConda.
- Start by downloading the version of MiniConda corresponding to your operating system from this page: [repo.anaconda.com/miniconda](https://repo.anaconda.com/miniconda/).
- The way you will have to install it depends on your operating system:

#### a. Windows

- What you downloaded should be a ".exe" installer.
- You can run it, it doesn't require you to be admin on your machine.
- When you will be asked where to install MiniConda, you can go with the default path, **BUT** __copy the path into notepad__ or  another editor and save it for later.
- From now on, whenever you will see "open a conda terminal", you should open your program named "Anaconda Prompt".

#### b. Mac/Linux

- What you downloaded should be a ".sh" file.
- Open a terminal (Ctrl + Alt + T on Linux or the application named "Terminal" on Mac).
- In the terminal, write `sh ` (with a blank space after `sh`)
- Drag and drop the ".sh" in the terminal: it should add its full path after the `sh `.
- You should have something like `sh /home/some/path/to/MiniConda3-latest-xxx.sh`.
- Press Enter and follow the instructions.
- Close everything and open a new terminal. If the installation is successfull, you should see `(base)` written on the left of the terminal's prompt.
- From now on, whenever you will see "open a conda terminal", you should simply open a new terminal.


### 2. Install Git

- Git is a version control system for source code.
- In our case, it will be useful to access some code hosted on GitHub.
- Start by going to the Git download page: [git-scm.com/install](https://git-scm.com/install/).
- If you indicate your operating system, you will be given the instructions to use.
- Once you executed the provided instructions, open a new conda terminal.
- Use the command `git --version`. If you are shown a version, everything went fine. Otherwise, Git installation failed.

**Warning:**

The `git` command will only be available in terminals opened after the installation is over. You should close all opened terminal to avoid running in troubles.

### 3. Napari

#### a. Create the environment

- Start by opening a new conda terminal.
- We will create a new environment in which we will install Napari.
- To create a new environment, use the command: `conda create -n napari-mldl-env -y python=3.12`
    - All conda commands start with `conda`.
    - `create` means that we want to create a new environment
    - `-n` means that the next thing that is coming in the command after the next blank space is the name that we want to give to our env.
    - `napari-mldl-env` is the name chosen for this env (it is arbitrary, you can use whatever you want).
    - `-y` means that the command will be executed without user-interaction, all prompts will be answered with yes.
    - `python=3.12` means that we don't let MiniConda choose which version of Python to install, we force it to use 3.12.

**Note**

If it is the first time ever that you use MiniConda, when you launch your first `conda create ...` the command will be interupted to ask you to accept the terms of services. Conda will show you the commands to accept them. Copy and paste **one after the other**. They should look something like `conda tos accept  ...`

You **must** relaunch the environment creation after accepting the TOS.

#### b. Install Napari

- Activate the environment that we just created with `conda activate napari-mldl-env`.
- Napari is available via pip.
- Install it with the command: `pip install "napari[all]"`.
- It should install it and all its dependencies.

#### c. Extra plugins

- Napari is just a 3D viewer, it doesn't have many features by itself. Most useful things are performed by plugins.
- Run the following commands in `napari-mldl-env`.
- Install the extra plugins with:
    - `pip install scikit-image scipy set-calibration matplotlib scikit-learn opencv-python-headless napari-trackastra napari-skimage`
    - `pip install git+https://github.com/ctrueden/yanse.git`


### 4. Cellpose

- We want to install Cellpose in a way that:
    - it can use your graphics card if you have an Nvidia one
    - it can be accessed as a module from Napari 
    - it can be accessed from other softwares
- Run the following commands in `napari-mldl-env`.
- Start by installing PyTorch with `pip3 install torch torchvision --index-url https://download.pytorch.org/whl/cu126`.
- Then install Cellpose with `pip install "cellpose[gui]"`.
- Eventually, install `cellpose-napari` with `pip install git+https://github.com/c-h-benedetti/cellpose-napari.git`.

### 5. Tensorflow

- Tensorflow is a deep-learning package.
- We will install it in an environment by itself to avoid collisions with PyTorch.
- Open a new conda terminal.
- Create a new environment with `conda create -n tf-env -y python=3.12`.
- Activate it with `conda activate tf-env`.
- Install everything required with `pip install tensorflow tifffile pandas notebook mpld3 openpyxl pydot scikit-learn scikit-image seaborn matplotlib`.

### 6. SAM server

#### a. Create the environment

- To use SAM, we will create an environment containing a server through which SAM's API is exposed.
- Start by opening a new conda terminal.
- Create a new environment with the command: `conda create -n samapi -y python=3.12`.
- Activate the environment with `conda activate samapi`.
- The `(base)` should have become `(samapi)` at this point.

#### b. Install SAM

- Start by installing PyTorch with the command: `pip install "torch==2.7.0" torchvision --index-url https://download.pytorch.org/whl/cu126`.
- If you are on **Windows**, you have a side dependency to install using the command: `pip install triton-windows`.
- You can now install SAM server from GitHub using Git and pip: `pip install git+https://github.com/ksugar/samapi.git`

## IV. QuPath

### 1. Itself

- You should download and install QuPath from [https://qupath.github.io/](https://qupath.github.io/).

### 2. CellPose

#### a. The plugin

- We need a way to get Cellpose (the Python modul) to communicate with QuPath (written in Java) so we can use it from QuPath.
- It requires a bridge between Cellpose and QuPath.
- Start by downloading the ZIP from the release list of [CellPose for QuPath](https://github.com/BIOP/qupath-extension-cellpose/releases/download/v0.12.1/qupath-extension-cellpose-0.12.1.zip).
- Unzip it wherever you want.
- Inside it, you should find a JAR and a PY. The PY is useless, we will only use the JAR.
- Open QuPath. Then, drag and drop the JAR in QuPath.
- If everything went fine, you should see Cellpose in the "Extension" menu from the top-bar.

#### b. Configuration

- We just installed the bridge.
- One end of the bridge is QuPath so there is not problem about that. The other end should be CellPose but we don't know where it is (yet).
- The first step is to locate the Python that CellPose is using. Doing that depends on the OS.
- On Mac/Linux: 
    - open a new conda terminal.
    - activate CellPose's env using `conda activate cellpose-env`.
    - launch the command: `which python`.
    - it should show you a path which is the one we are interested in.
- On Windows:
    - when you installed MiniConda, you were asked to write down where it is installed.
    - open the folder in which you installed MiniConda.
    - In there, you should have an `envs` folder.
    - Within `envs`, you should have a `cellpose-env` folder.
    - Within the `cellpose-env` folder, you should have a "python.exe" file.
    - It's the path of this file (including the ".exe") that we want.


**Warning**

Be careful, by default, file extensions are hidden on Windows: you won't see the ".exe" in Python's file. On Mac and Linux, binaries don't have any particular extensions.

- In QuPath, in the top-bar menu, go to "Edit" > "Preferences".
- You should have a "CellPose/Omnipose" section in there.
- In the `CellPose SAM 'python.exe' location`, paste the path of Python.
    - On Windows: you should replace `\` with `/` and have "python.exe" not just "python".

### 3. SAM

- Download the JAR corresponding to the [SAM extension](https://github.com/ksugar/qupath-extension-sam/releases/download/v0.9.1/qupath-extension-sam-0.9.1.jar).
- Drag and drop the JAR into QuPath.
- If everything went well, you should see it in the "Extensions" menu.
- This extension is not transparent, each time **before launching SAM in QuPath** you need to:
    - open a new conda terminal.
    - activate SAM API env with `conda activate samapi`.
    - launch the server with `uvicorn samapi.main:app --workers 2`.
    - it will download the models (15 GB) so let it run for a while.

## V. Fiji

### 1. Itself

- Even if you already have a Fiji installed on your computer, you should read this section and install a new Fiji.
- Start by going on [Fiji's download page](https://imagej.net/software/fiji/downloads) and get the one corresponding to your OS from the **Latest** section, not the **Stable** section.
- Unzip it and keep it in your documents, or create an "applications" folder in your home repository but don't move it in "Program Files" or "Programs x86" or any of these folders.

### 2. Extra plugins

- Launch Fiji from the binary inside the folder.
- Go to ""Help" > "Update...". At this point, Fiji will search for updates. When it will be done, you will have a new "ImageJ Updater" window.
- Go to "Manage Update Sites" in the lower left corner.
- Select the following update sites:
    - `ImageScience`
    - `IJPB-plugins`
    - `LabKit`
    - `SAMJ`
    - `clij`
    - `clij2`
- Once you selected them all, click on "Apply and Close" and then on "Apply Changes".
- The new plugins will be downloaded.

### 3. SAM's model

- Re-open Fiji if you just installed some new plugins.
- Go to "Plugins" > "SAMJ" > "SAMJ Annotator".
- Click on the triangle next to "SAM2 Tiny".
- In the extension that just appeared, click on "Install" to download the required model. It can take a while.

## V. VSCode

- Download and install VSCode from [https://code.visualstudio.com/download](https://code.visualstudio.com/download).
- Open VSCode.
- In the left column on the window, you should see a column of icons of top of each other.
- There should be one looking like 4 squares with the upper-right square being tilted. It should say "Extensions" when you pass your mouse over it.
- Click on it. It should open a new panel with a search box on top saying "Search Extensions in Marketplace".
- In the search bar, type "Python". In the list of results, install the one simply named "Python" made by Microsoft. It should have a little green angle in the upper-left part of its icon.
- Of the same way, search for "Jupyter" and install the one simply named "Jupyter".
