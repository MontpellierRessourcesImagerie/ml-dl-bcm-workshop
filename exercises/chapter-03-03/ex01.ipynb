# 3.3 Segment Anything Model (SAM)

## 3.3.1 Segment mitochondria in EM images

#### Required data:

| **Folder** | Description | Location | License |
| --- | --- | --- | --- |
| Electron Microscopy Dataset | EM slices containing mitochondrias | [Kaggle](https://www.kaggle.com/datasets/dat0chin/electron-microscopy-dataset) | CDLA-Sharing-1.0 |

#### Goals:

Eletron microscopy images are hard to work with since all structures are present at once. In this context, finding a precise object can be hard. Specialized deep-learning models exist depending on what you are looking for but they may not be designed for your object of interest or not suitable for your images. 

In this case, manually annotating your images is your last resort. However there is a something between fully-manual and fully-automated segmentation that would be semi-automatic (or assisted) segmentation.

In our case, it will consist in feeding the model with some easy to produce inputs (bounding boxes, points, scribbles, ...), and getting the precise segmentation as the result of inference.

### 3.3.1.1 Using SAM from QuPath

#### A. Goals

- Our goal will be to segment mitochondrias in QuPath using SAM.
- We want to have an individual polygon for each instance of mitochondria.
- By the end, we would like to produce a CSV file with one line per instance giving its circularity, its area and the image it comes from.

#### B. Launch SAM's server

- Open a new conda terminal and activate the environment containing SAM's server with: `conda activate samapi`.
- Launch the server with the command: `uvicorn samapi.main:app --workers 2`.
- If you have never launched the server before, it will download all the models (approx. 15GB).
- Since this is a server, the command will never finish, the process will keep running and we will communicate with it via a client in QuPath.

#### C. Segment the mitonchondrias

- Start by creating a new project in QuPath and import some images inside.
- Set the image type to "Fluorescence". It is not the actual type, but a non-RGB image is fluorescence for QuPath.
- In the "Annotation" tab of QuPath, next to "Class list", use the "+" button to create a new class that you will name `Mitochondria`.
- SAM can work only with rectangle and point annotations. In this exercice, we will only work with rectangles.
- Make a rectangle around each mitonchrondria that you see.

**Tip**

You can quickly place all your rectangles by using the keyboard shortcut [R]. Whenever you make a new rectangle, QuPath sets you back to panning mode. Simply keep a finger on [R] and the other on the left-click! This way, you can place rectangles and move afterward.

- In the top-bar menu, go to "Objects" > "Select" > "Select annotations". It shoud select every rectangle that you made.
- Now, find the `Mitochondria` class that you created in the "Class list" section and click on it.
- If you now click on "Set selected" (just below the list), this class will be given to all the rectangles you made.

#### D. Run SAM inference

- If you go to "Extension" > "SAM" (located in the top-bar menu), a new floating window should appear.
- Make sure that all your rectangles are still selected in QuPath. If it is the case, they should be drawn in yellow in the viewer.
- In SAM's window, make sure that "Assign random colors", "Keep prompts", "Assign names" and "Display names" are unchecked.


**Warning:**

SAM works only on what is **currently visible** in the viewer at the viewer's **current resolution**. Don't forget to zoom-out just enough for all your rectangles to be visible your viewer.

- You can now click on "Run for selected".
- A little spinning wheel should appear in the upper-right corner while the inference is running.
- After the inference is completed, to match our initial goals, we need to include the measurements of area and circularity for each object.
- To do so, make sure all the mitochondria are still active selections (yellow in the viewer) and go to "Analyze" > "Calculate feature" > "Shape feature".
- You can keep everything or only keep the area and the circularity, the processing of these properties is nearly instantaneous.

#### E. Measurements

- Repeat the operation for another image (or more if you wish).
- Once you are done, save your project with [Ctrl]+[S] or "File" > "Save".
- Go to "Measure" > "Export measurements...".
- Move to the right column all the images on which you performed the segmentation.
- Select a path the output file (the TSV) that will contain our measurements.
- In "Export type", you must say what type of object holds the valuable information. In our case, annotations do (each polygon we made is an annotation).
- If you click on "Populate", the dropdown menu will be filled with all the available columns that can be injected in the TSV. To avoid  it with useless data, select only what you are interested in.
- You can now click on "Export" and use the produced TSV file in your favorite spread sheet software.
