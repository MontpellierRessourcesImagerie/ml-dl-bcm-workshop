### 3.3.2.1 SAMJ from ImageJ

#### A. Goals

- We will use the SAMJ plugin from ImageJ to create ROIs around our C. Elegans just by providing it with bounding-boxes.
- This plugin encodes the image and keeps the latent space of the model loaded, only the decoding is performed when we make a box.
- ROIs around worms will be added to the manager.
- It is interesting because it allows us to deal with worms on top of each other.
- Eventually, we will use a Jython script to turn these ROIs into polylines through the process of skeletonization so we can measure their length.


#### B. Launch the GUI

- Open the image on which you want to work.
- If the RoiManager window is opened, close it.
- In Fiji's search bar, type "SAMJ" or go to "Plugins" > "SAMJ" > "SAMJ Annotator".
- A new floating window should have appeared.
- If you didn't initialize it at explained [in the requirements](../../exercises/extras/install-requirements.html), you should do it now.
- Select the model that you want to use (`SAM2 Tiny` or `SAM2 Small`).
- In the dropdown menu, select the image that you just opened.
- If you now click on "Go", the image is going to be encoded, and the "OFF" button should turn red.

**Warning:**

Models are quite heavy, they require a good internet connection to be downloaded.

#### C. Annotate your image

- Once you are ready to annotate, pass your selection tool to "Rectangle" and click on the red "OFF" button. It should turn green and say "LIVE".
- Now, every time you will draw an AABB on your image, it will be fed to SAM for inference.
- The process can take 1 or 2 seconds, but your rectangle should disappear and a polygon ROI should appear both on your image and in the manager.
- The boxes should be as close as possible of the target object. It doesn't matter if a big chunk of another object is included in your box. It will happen a lot with worms at an angle of 45° for example.

**Tip:**

If SAMJ fails at stoping at the correct place (goes in the neighboring worm) you can:
- select the faulty ROI in the manager by clicking on it
- switch to the lasso selection (the button in shape of bean)
- With the [Alt] key pressed, make a lasso around the wrong piece of the ROI
The modification will be applied into the manager as well. You can switch back to rectangles to continue.

#### D. Turn polygons into skeletons

- We will manually turn one of the polygons into a skeleton to see the way it is done, but we will later use a script to do so.
- In the RoiManager, click on any of the rois that you would like to skeletonize.
- Do a right-click on the image and go with "Duplicate...". Make sure that "Ignore selection" is checked and validate. A new identical image with the roi should appear.
- From Fiji's search bar, you will search an run the following commands:
    - "8-bit"
    - "Clear Outside"
    - "Fill"
- You can now remove the roi with either [Ctrl]+[Shift]+[A] or "Edit" > "Selection" > "Select None".
- Again, from Fiji's search bar, you will run "Skeletonize (2D/3D)"

**Warning:**

Fiji embeds both an old skeletonization algorithm called "Skeletonize" and the new called "Skeletonize (2D/3D)". Be carefull not to go with the old one that can produce undesirable extra branches.

- You can now launch the "Analyze skeleton (2D/3D)". You must adjust the settings:
    - `prune cycle`: none
    - `prune ends`: disabled
    - `calculate longest shortest path`: yes
    - `show detailed info`: as you wish but preferably disabled
    - `show labeled skeleton`: as you wish but preferably disabled
- A new window called "Longest shortest path" should have appeared, in which the longest path is in magenta.
- If you didn't have undesirable branches, you should have only this path. However, if you had garbage branches, this process helps you removing them.

- Doing so for all worms would take forever, so [here is a Jython script](./make_skeletons.html) allowing you to do it for the whole roi manager.
- You can open it in Fiji and run it right away, there is no setting to adjust, it just needs the original image and the manager.

#### E. Measure lengths

- Once the manager is filled with the polylines, click on the "Deselect" button in the right column, next to the list of roi names.
- You can now click on "Measure".
- It should produce a new results table with one line per worm and a "Length" column.
- Images are not calibrated (the physical pixel size is not set) so the lengths are in pixels.
