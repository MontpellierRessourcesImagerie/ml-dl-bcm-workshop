# Translocation measurement script

This Groovy script loops over cells in an image and:
- Finds the median green signal in each cytoplasm and each nucleus
- Processes the translocation ratio
- Inject the value in the cell's measurements list

You can simply open it in QuPath and run it for the project.

⚠️ In the script editor, in the top-bar menu, verify that "Run" > "Include default imports" is activated.

---

```groovy
selectAnnotations()
removeObjectsTouchingSelectedBounds()

selectDetections()
objs = getSelectedObjects()
q90_factor = 1.28

for (o: objs) {
    m_list = o.getMeasurementList()
    
    med_nucleus = m_list.get("Nucleus: Green: Median")
    med_cyto    = m_list.get("Cytoplasm: Green: Median")
    std_nucleus = m_list.get("Nucleus: Green: Std.Dev.")
    std_cyto    = m_list.get("Cytoplasm: Green: Std.Dev.")
    
    // Q90 approximation under normality
    i_nucleus = Math.max(0, med_nucleus + q90_factor * std_nucleus)
    i_cyto    = Math.max(0, med_cyto    + q90_factor * std_cyto)
    
    r = (i_nucleus - i_cyto) / (i_nucleus + i_cyto)
    m_list.put("Translocation ratio", r)
}
```
