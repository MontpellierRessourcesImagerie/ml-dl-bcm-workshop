### 3.2.2.1 Using Napari and Python

#### A. Goals

- We are going to use Napari again to call Cellpose but this time, to work on a 3D image.
- Napari will help us deal with the anisotropy that's often hard to deal with when using Cellpose.
- Afterward, detecting spots is a rather simple process, so we will do it in Python from inside Napari.
- The last step will be to find a way to assign each spot to the nucleus owning it.

#### B. Setup the workspace

- Open a new conda terminal and activate the environment in which you installed Cellpose and Napari with the command `conda activate cp-env`.
- Launch Napari by typing `napari` and then press [Enter].
- From the "Plugins" > "Calibration Tool" menu, open "Scale Tool" and "Split Channels".
- Open the "spots-in-nuclei.tif" image in the viewer.
- If you are asked what should Napari use to open image, go with "napari-builtins".
- Use the "Calibration Tool" to set the physical size of voxels:
    - `0.6031218` for "X" and "Y".
    - `0.981132` for "Z".
    - `µm` as unit.
    - `ZCYX` for "Axes" (you can deduce it from the second axis having a size of 2, which is the number of channels).
- You can apply these settings.
- You can move to the "Split Channels" widget and run it right away.
- If you use the slider below the viewer, you should see your two channels (magenta and green).

#### C. Segment the nuclei

- To make up some room, you can close both the widgets from the "Calibration Tool".
- Then, from "Plugins" > "cellpose-napari", open the "CellPose Inference" widget.
- Set the "Main channel" to the layer containing your nuclei. We don't have any secondary channel to help us.
- In 2D view, adjust the "Median diameter" of objects. It should be somewhere between 20 and 25 pixels.
- To remove potential oversegmented objects, increase the "Minimal object size" to 20 or 30 pixels.
- You can now launch the segmentation. 

**Warning:**

Segmenting a 3D image is both memory expensive and computing expensive. If your computer cannot do it, you will find the segmented result in the data for this exercise.

**Note:**

Napari's progress bar won't help you much, you can look at the terminal to know the segmentation progress. In 3D, Cellpose has to work on all 3 planes before "stitching" the result, even with a good PC and a GPU it is a long operation.

#### D. Detect spots

- Once the segmentation is done, you can close Cellpose's widget to free some space.
- To detect spots, we will perform an opening by reconstruction on our spots channel and search the local maxima exceeding a certain prominence in the resulting image.
- The opening by reconstruction and the search for maxima given a prominence are already implemented in Python in the `scikit-image` package, we don't have to implement them ourselves.
- We defined arbitrarily that for a maxima to be a spot and not noise, it must have a prominence at least equal to 1.5 times the standard deviation in the whole image (you can tune it later if you wish).
- From the "Plugins" menu, open the "Script Editor". By default it is docked like any other widget, buy you can undock it by double-clicking on the widget's name above it once it's opened.
- You can find [here the script that we are going to use](./napari_detect_spots.html). Each block of code is explained, you can transfer it chunk by chunk into your script editor and modify what must be adjusted.
- Once everything is in the script editor, you can launch the script. It should take up to a minute depending on your computer.
- The spots should be visible in your viewer.
- If you click on the layer containing the segmented nuclei and then go to "Plugins" > "Features table widget", you should be able to visualize the array of counters that we created from the script.
- You can save
