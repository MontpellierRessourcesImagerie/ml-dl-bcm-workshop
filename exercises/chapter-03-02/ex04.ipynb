### 3.2.4.1 Batch segmentation from a Python notebook

#### A. Goals

- We will try to achieve our objective of counting cells on each image automatically, without dealing with images one by one.
- The Cellpose basic operations (inference, batch, training) are available through Python workers implemented in `cellpose-napari`.
- These workers are exposed via Napari interface, but they can be called from anywhere in Python.
- To be able to blend some code and corresponding explanations, we will try to do everything in a Python notebook.


#### B. Getting started

- The notebook is available in the `notebooks/chapter-03-02/batch_cellpose.ipynb`.
- The instructions are mixed with the code in there.
