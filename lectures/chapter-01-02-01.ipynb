Clustering can be used as an unsupervised machine learning method for classification tasks. Clustering means that objects are grouped in a way, that objects in the same group are more similar to each other than to objects in other groups. A large number of clustering algorithms and variants exist, for example ``k-means``, ``DBScan``, ``Hierarchical clustering``, ``Expectation maximization``. We will have a closer look at the ``k-means clustering`` here.

## 2.1 k-means clustering

The k-means clustering partitions the feature space into k clusters. Each feature vector belongs to the cluster with the nearest mean. The basic algorithm is very simple.

<center>Algorithm 1 - k-means clustering</center>

```
1. Start with k initial clusters
2. Repeat until convergence
   2.1 Assign feature vectors to clusters
   2.2 Recalculate the means of the clusters
```

You can step through the algorithm in the [k-means visualization](https://www.naftaliharris.com/blog/visualizing-k-means-clustering/) from Naftali Harris. Try different initialization strategies and different datasets. You can also try the [DBScan visualization](https://www.naftaliharris.com/blog/visualizing-dbscan-clustering/) and compare. DBScan is a density based clustering algorithm.

One difficulty with the k-means algorithm is, that have to know the number of clusters beforehand. The k-means algorithm always converges, however the result and the conversion speed depend on the initial means. A simple initialization strategy is to choose the first mean randomly and then to choose each next mean as the data point the furthest away from the previous mean. A more complex initialization strategy is called k-means++. The first mean is choosen randomly. Each next mean is choosen randomly with a probability proportional to the square of its distance to the previous mean.

In some situations the k-means algorithm will not find the clusters we probably expect. This is for example the case when the mean is not within the cluster, but far away. This is the case for the ``Smiley Face`` example in the visualization.

## 2.2 k-means clustering as machine learning

How can we use the k-means algorithm as a machine learning algorithm? Say we want to cluster the pixels in an image based on the values of some feature vectors. We can train the k-means by selecting a small percentage of the pixels and running the k-means on the corresponding feature vectors. The means obtained in the training phase are the trained model. To predict the class of a new pixel, we calculate its feature vector and find the closest mean to it.

## 2.3 Color clustering

Simple features that we can use in color images are the color components of a pixel. The rgb values would be one option, but using different color spaces can be interesting. The CIEL*a*b* color-space has the axes L* for lightness, a* which goes from green in the negative values to red in the positive values and b* from yellow to blue. It is designed so that the distances between colors in the color-space correspond to distances perceived by humans.

However, different features have different value ranges. For the k-means algorithms, this is a problem, because we calculate the Euclidean distance between vectors. If one component always has very small values and another always very large values, the distances will be dominated by the large values and the feature with the small values will be ignored.

To avoid the problem of too different value ranges, we can apply a standardization or z-score normalization to the feature values $f_k$ of a feature. The z-score normalization shifts the mean $\mu$ of the distribution to zero and normalizes the standard deviation $\sigma$ to one.


$$
z(f_k) = \frac{f_k - \mu}{\sigma}  ~~~~~~~~~~~~ (1)
$$
