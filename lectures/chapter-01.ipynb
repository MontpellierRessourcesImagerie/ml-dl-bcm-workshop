### 1.1.1 What is bio-image analysis?

In [[1]](#lit-ref-1) image analysis is defined as:

<cite>"Image analysis is concerned with making quantitative measurements from an image
to produce a description of it."</cite>

In analogy we define bio-image analysis as:

<cite>"Bio-image analysis is the extraction of information from digital images in the context of biological research."</cite>

<center>
<img width='55%' src="../media/chapter01_01/analysis.png" alt="analysis" >
<br>
<small>
Fig. 1.2 &ndash; Image analysis is the extraction of quantitative data from digital images.
</small>
</center>

As is illustrated in fig. 1.3, an image analysis workflow can usually be divided into different phases: pre-processing, segmentation and post-processing. In the pre-processing phase a number of filters are typically applied to the image to make the structures of interest stand out from the background. The pre-processed image is then segmented to obtain a binary mask separating the background pixels from the foreground pixels or a label image, if multiple objects need to be segmented individually. The mask is then post-processed to correct segmentation errors, often using mathematical morphology [[2]](#lit-ref-2). The object or the objects defined by the post-processed mask or label image are then measured on the mask/label image and the original intensity image, to extract the data from the image.

<center>
<img width='75%' src="../media/chapter01_01/workflow.png" alt="workflow" >
<br>
<small>
Fig. 1.3 &ndash; A typical image analysis workflow.
</small>
</center>

One or more, but usually a small number of features are often extracted, by applying filters to the original image. The aim is to obtain features that can be combined to an image on which the segmentation is easier to accomplish than on the original image. Before the features are extracted, a scale is selected by applying a Gaussian Filter with a given sigma.

The Gaussian filter replaces the value of a pixel by a weighted average of the pixels in a neighborhood around the pixel. The weights follow a Gaussian function. They become smaller with the distance from the central pixel. [[3]](#lit-ref-3)

The Gaussian Filter is a low pass filter that will remove the high frequencies. The high frequencies represent the noise and small objects in the image. With larger sigma only even lower frequencies pass.

</small>
</div>

One kind of filters frequently used are ``convolution filters`` [[4]](#lit-ref-4). Convolution filters are linear filters. A convolution filter in digital image processing can be represented by a matrix with a radius $r$, called the kernel of the filter. To apply the filter to an image, the central position of the kernel is iteratively placed on each pixel of the image. A new value for that pixel is calculated, based on the values in the original image. It is calculated as the weighted sum of the values in the neighborhood with radius $r$, where the weights are the values in the kernel.

It should be clear, that the Gaussian filter is a convolution filter, but there are others, for example the Mean filter, the Laplacian filter and the Sobel filter which are used for spot and edge detection.

A different class of filters are rank filters [[5]](#lit-ref-5). Rank filters sort the values in the neighborhood and answer the value at a specific position, for example the first last or the one in the middle. Examples are the Min, Max and Median filters. In contrast to convolution filters, rank filters are not linear filters.

<center>
<img width='35%' src="../media/chapter01_01/2D_Convolution_Animation.gif" alt="convolution" >
<br>
<small>
Fig. 1.6 &ndash; Animation explaining the discerte 2D deconvolution.<br>(By Michael Plotke - Own work, CC BY-SA 3.0, https://commons.wikimedia.org/w/index.php?curid=24288958)

</center>

Mathematically we can express a discrete convolution of a digital 2D image $I$ with a kernel $k$ of radius $r$ as:

$$
    (I*k)(x,y) = \sum_{m=-r}^{r}\sum_{n=-r}^{r}I[x-m,y-n] \cdot k[m,n] ~~~~~~~~~~~~ (1)
$$

### 1.1.2 Machine Learning [[7]](#lit-ref-7)





<cite>"Machine learning algorithms build a mathematical model based on sample data, known as ”training data”, in order to make predictions or decisions without being explicitly programmed to perform the task."</cite> [[6]](#lit-ref-6)





<center>
<img width='55%' src="../media/chapter01_01/model.png" alt="model" >
<br>
<small>
Fig. 1.8 &ndash; A mathematical model maps input values to output values by applying functions, depending on parameters. The parameters can be modified to obtain a specific mapping.
</small>
</center>

To do machine learning, you first decide which kind of model you want to use. The model could be a mathematical function, a random forest, a neural network, or anything else that maps input values to unique output values and has parameters that can be ``tuned``.

<img width='15%' src="../media/chapter01_01/training.png" alt="training" >

In the ``training`` phase the model is learned from the training data, by modifying the parameters of the model. After the training phase follows the validation phase.

<img width='8%' src="../media/chapter01_01/validation.png" alt="validation" >

In the ``validation`` phase the model is evaluated on test data that has not been used in the training phase. When its performance is satisfactory the application phase can begin.

<img width='8%' src="../media/chapter01_01/application.png" alt="application" >

In the ``application`` phase, the trained model is used to make predictions on new data.

We distinguish between supervised and unsupervised training and between classification and regression models.

<img width='8%' src="../media/chapter01_01/supervised.png" alt="superwised" >

The training is called ``supervised``, when the model is trained from pairs of input and output data.

<img width='8%' src="../media/chapter01_01/unsupervised.png" alt="unsuperwised" >

The training is called ``unsupervised``, when the model is trained on the inherent structure of the input data alone.

<img width='8%' src="../media/chapter01_01/classification.png" alt="classification" >

The model is called a ``classification``-model, when the output is one of a small number of categories.

<img width='18%' src="../media/chapter01_01/regression.png" alt="regression" >

The model is called a ``regression``-model, when the output is a real number.

When we talk about machine learning in this workshop, what we mean is that we are trying to find the parameters of a mathematical model, that minimize the overall prediction error.

### 1.1.3 The Linear Regression Example

We will now study a very simple example of machine learning. We use a simple example, so that we can concentrate on the machine learning aspects first, without having to understand a complex model in the same time. The model is so simple that out off the educational context it would not make sense to use machine learning for it at all, since a simple analytical solution exists.

Consider the following scenario. Archeologists study the bones of an ancient populations of humans. Sometimes whole skeletons are found, but most of the time the individual skeletons can not be reconstructed from the bones found. Let's say the archeologists are very interested in the body height of the population in question, and we know that there is a linear relation between the length of the femur bone and the body height. If the archeologists had a way to predict the body height from the length of the femur bone, they could obtain a much larger database about the body height of subpopulations.

Since we assume a linear relationship between the femur length and the body height, we take a simple linear function as our mathematical model.

$$
    f(x) = \omega_1 + \omega_2 \times x  ~~~~~~~~~~~~ (2)
$$

$f(x)$ is the body height and $x$ is the femur length. The y-intercept $\omega_1$ and the slope $\omega_2$, are the parameters of the model. The training data are the pairs of femur length and known body height in the table in figure 1.10.

Our task is now to find parameter $\omega_1$ and $\omega_2$, so that the error between the training data and the predictions from the model is minimal. If we look at one pair of input (femur length) and output data (body height), and one fixed set of parameters $\omega_1$ and $\omega_2$, we can calculate the loss as a function of the difference between the predicted output and the ground truth output. We could use the absolute difference, but the squared difference has some advantages for our scenario.

$$
    L(x_i;\omega_1, \omega_2) = \left( y_i - f(x_i; \omega_1, \omega_2) \right)^2  ~~~~~~~~~~~~ (3)
$$

What we need to minimize now, the cost function corresponding to our loss function, is the average loss on the whole training dataset. With the squared difference as loss, we obtain the mean squared error as a cost function.

$$
    C(\omega_1, \omega_2) = \frac{1}{N} \sum_{i=1}^{N} \left(y_i - f(x_i; \omega_1, \omega_2)\right)^2 ~~~~~~~~~~~~ (4)
$$

</small>
</div>
<div style="width:45%; float:left;  ">
<img src="../media/chapter01_01/cost_plot.png" alt="cost plot" >

Fig. 1.12 &ndash; The cost function $C(\omega_1, \omega_2)$.

We can now start with arbitrary values for $\omega_1$ and $\omega_2$ and use an optimization algorithm to find values that minimize the cost function. One such algorithm is called Gradient Descent. In each step it advances into the direction of the steepest descent (the opposite of the gradient) with a given step width. The step width can be variable. As an analogy, imagine you are on a mountain top, covered by thick mist. You want to descent from the mountain as quick as possible, however you can not see in which direction the valley lies, because of the mist. To nevertheless get quickly to the valley, you take each step in the direction with the steepest descent.

</small>
</center>

You can follow a gradient descent on the problem yourself, by executing the script [linear_regression.ijm](../scripts/linear_regression.ijm). Run the script and set a start point on the cost function. Press ok to execute one step the gradient descent. If you want to go faster, you can hold down ``alt``.

After we found the optimal parameter values, or at least parameter values with a reasonable small cost, we have a trained model, that we can use to make predictions. Let our trained model be:

$$
   f(x) = \omega_1 + \omega_2 \times x \mathrm{~~~with~~~}  \omega_1 = 131.13cm \mathrm{~~~and~~~}  \omega_2 = 0.878 ~~~~~~~~~~~~ (5)
$$



We use the model to predict the body height for a femur bone with a length of 55cm we found:



$$
f(55cm) = 131.13cm + 0.878 \times 55cm = 179.42cm
$$

### 1.1.4 Counting spots without machine learning

We want to develop a workflow to automatically count the number of spots in images as shown in figure 1.15. We have ground truth images available, that allow us to know how close our result comes to the correct number.

<center>
<img width='65%' src="../media/chapter01_01/spots.png" alt="spots" >
<br>
<small>
Fig. 1.16 &ndash; The aim is to automatically count the spots in these kind of images.
</small>
</center>

A very simple approach would be to threshold the image to separate the foreground from the background and then to apply a binary watershed (distance map watershed) operation, to separate the objects from each other. To then count the spots, we calculate the connected components and get the maximum value in the label image.

<center>Macro 1 - A first attempt to count the spots</center>

```java
setOption("BlackBackground", true);
setAutoThreshold("Yen dark");
run("Convert to Mask");
run("Watershed");
run("Connected Components Labeling",
    "connectivity=4 type=[16 bits]");
count = getValue("Max");
print("Nr. of spots: " + count);
```

The result we obtain is 1420 spots are counted. The corresponding ground-truth image contains 1947 spots. If we apply the squared difference loss, we obtain a loss of 277729.

A more experienced image analyst, might come up with a more complex solution, taking care of the gradient in the background, using a LoG filter to detect the spots.  

<center>Macro 2 - A more sophisticated attempt to count the spots.</center>

```java
inputImageTitle = getTitle();
run("Duplicate...", " ");
run("Gaussian Blur...", "sigma=36");
backgroundTitle = getTitle();
imageCalculator("Subtract create", inputImageTitle, backgroundTitle);
run("FeatureJ Laplacian", "compute smoothing=1");
run("Find Maxima...", "prominence=8 light output=[Single Points]");
run("Connected Components Labeling", "connectivity=4 type=[16 bits]");
count = getValue("Max");
print("Nr. of spots: ");
```

This time we count 1544 spots and the loss becomes 162409. That is better than before, but still far from the correct value 1947.

### 1.1.5 Introduction to labkit [[8]](#lit-ref-8)

We will now use a machine learning method called random forest. The method will be explained in chapter 3. In this case the model consists of a set of decision trees (a forest). The individual decisions concern the values of the features extracted from the input image.

We will use the FIJI/ImageJ-plugin ``labkit``. Labkit can classify pixels in 2D and 3D images, using an arbitrary number of classes. It can be used on big images, that are bigger than the available RAM. It can use the graphics board to speed up calculations, when the clij2-plugin is installed. A trained classifier can be saved and applied to images. Labkit also offers batch processing. A classifier can be applied to all images in a given folder.

We want to train a classifier for the spot images. In a dataset there will usually be some variation, some images might be brighter than others, some might contain additional biological objects not present in others and so on. We should train the classifier on a representative subset of the data, so that the classifier will reliably work on the whole dataset and other similar images.

##### **1.1.5.4 Training a classifier**

By default, there are the labels ``background`` and ``foreground``. We add another label ``separations``, that we will use to separate touching spots. Normally the pixel classification only does semantic segmentation, affecting a class label to each pixel in the image. By adding a class for separations and running a connected components labeling on our foreground label, we will obtain an instance segmentation, in which each object has its own label. To add the additional label, press the ``Add label``-button. Rename the label by double-clicking on it and entering the new name. The small triangle button next to each label opens a context menu. You can change the color of the label by clicking on the color field left of the label's name.

We will now use the drawing tool to select some example pixels for each of the classes. Please note that drawing the whole objects would be a bad strategy, since we would add a lot of very similar values to the foreground class. Labkit's user guide advices to draw one pixel wide lines from the centers of the objects to the borders for the foreground. For the background you should in a similar way draw one pixel wide lines from the border away from the object. However sometimes it can be better to do even more sparse annotations, using only isolated points. The way to work is to start with a small number of annotations for each class, train the classifier and display the result. Now add some more annotations at places where there are still errors, train again, and so on, until the result is acceptable on all training images.

To add annotations for a class, select the corresponding label in the list. Either activate the draw tool and draw the annotations onto the image (left mouse click and drag), or hold down the ``d``-key while drawing. Make sure that you annotate a similar number of pixels for each class. When you make a mistake, use the eraser tool to correct it. Instead of activating the tool you can hold down the ``e``-key. You can use the slider at the upper-right to adjust the radius of the drawing and the eraser tool. Once you made some annotations for each class, train the classifier. You can use the triangle button next to the classifier or the ``Train CLassifier``-command from the ``Segmentation``-menu (``ctrl+shift+T``).
