## 1.4 Counting spots without machine learning

We want to develop a workflow to automatically count the number of spots in images as shown in figure 1.15. We have ground truth images available, that allow us to know how close our result comes to the correct number.

<center>
<img width='65%' src="../media/chapter01_01/spots.png" alt="spots" >
<br>
<small>
Fig. 1.16 &ndash; The aim is to automatically count the spots in these kind of images.
</small>
</center>

A very simple approach would be to threshold the image to separate the foreground from the background and then to apply a binary watershed (distance map watershed) operation, to separate the objects from each other. To then count the spots, we calculate the connected components and get the maximum value in the label image.

<center>Macro 1 - A first attempt to count the spots</center>

```java
setOption("BlackBackground", true);
setAutoThreshold("Yen dark");
run("Convert to Mask");
run("Watershed");
run("Connected Components Labeling",
    "connectivity=4 type=[16 bits]");
count = getValue("Max");
print("Nr. of spots: " + count);
```

The result we obtain is 1420 spots are counted. The corresponding ground-truth image contains 1947 spots. If we apply the squared difference loss, we obtain a loss of 277729.

A more experienced image analyst, might come up with a more complex solution, taking care of the gradient in the background, using a LoG filter to detect the spots.  

<center>Macro 2 - A more sophisticated attempt to count the spots.</center>

```java
inputImageTitle = getTitle();
run("Duplicate...", " ");
run("Gaussian Blur...", "sigma=36");
backgroundTitle = getTitle();
imageCalculator("Subtract create", inputImageTitle, backgroundTitle);
run("FeatureJ Laplacian", "compute smoothing=1");
run("Find Maxima...", "prominence=8 light output=[Single Points]");
run("Connected Components Labeling", "connectivity=4 type=[16 bits]");
count = getValue("Max");
print("Nr. of spots: ");
```

This time we count 1544 spots and the loss becomes 162409. That is better than before, but still far from the correct value 1947.
